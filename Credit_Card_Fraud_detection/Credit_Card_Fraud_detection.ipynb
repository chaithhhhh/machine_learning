```python
legit_sample=legit.sample(n=492)
legit_sample
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset["Class"].value_counts()
```

```python
x=new_dataset.drop(columns="Class",axis=1)
y=new_dataset['Class']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy score on Training data : ",training_data_accuracy)
```

```python
#Accuracy on testing data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy score on Testing data : ",testing_data_accuracy)
```

```python
def predict_transaction(model, transaction_data):

    prediction = model.predict(transaction_data)
    return prediction
```

```python
def predict_transaction(transaction_data):
    transaction_data = [float(val) for val in transaction_data.split(",")]  # Convert the input string to a list of floats
    prediction = model.predict([transaction_data])  # Model expects a 2D array, so wrap the data in a list
    return "fraudulent" if prediction[0] == 1 else "Healthy"

iface = gr.Interface(
    fn=predict_transaction,
    inputs="text",  # Input as a comma-separated string of feature values
    outputs="text",  # Output as a text label ("Normal" or "Fraudulent")
    live=False,
    title="Credit Card Fraud Detection",
    description="Enter transaction features as a comma-separated string.",
)

iface.launch()
```
